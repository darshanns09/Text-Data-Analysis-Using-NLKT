```python
import nltk
import string

# downloading stopwords list
nltk.download('stopwords')

# path of stopword folder
stopwords_folder = 'StopWords'

# empty set
stop_words = set()

# combining stopword by looping textfile in folder
import os

for file_name in os.listdir(stopwords_folder):
    if file_name.endswith('.txt'):
        file_path = os.path.join(stopwords_folder, file_name)
        with open(file_path, 'r', encoding='ISO-8859-1') as stop_words_file:
            stop_words.update(stop_words_file.read().splitlines())

# Loading Master dictionary
positive_words = set()
negative_words = set()

with open('positive-words.txt', 'r', encoding='utf-8') as positive_words_file:
    positive_words = set(positive_words_file.read().splitlines())

with open('negative-words.txt', 'r', encoding='ISO-8859-1') as negative_words_file:
    negative_words = set(negative_words_file.read().splitlines())
    
# Loading text from the output.txt file
with open(TextFile, 'r', encoding='utf-8') as file:
    text = file.read()

# tokenize the text into word
words = nltk.word_tokenize(text)

# remove stop words and punctuation
cleaned_words = [word for word in words if word.lower() not in stop_words and word not in string.punctuation]

positive_score = sum(1 for word in cleaned_words if word in positive_words)
negative_score = sum(1 for word in cleaned_words if word in negative_words)

polarity_score = (positive_score - negative_score) / (positive_score + negative_score + 0.000001)
subjectivity_score = (positive_score + negative_score) / (len(cleaned_words) + 0.000001)

print(f'Positive Score: {positive_score}')
print(f'Negative Score: {negative_score}')
print(f'Polarity Score: {polarity_score}')
print(f'Subjectivity Score: {subjectivity_score}')
```

## 2.    Analysis of Readability

AVG SENTENCE LENGTH 

PERCENTAGE OF COMPLEX WORDS 

FOG INDEX 

```python

# tokenize the text into word
sentences = text.split('.') 
words = text.split()  # Split text into words

average_sentence_length = len(words) / len(sentences)

# assume complex words are words with more than 4 characters.
complex_words = [word for word in words if len(word) > 4]

percentage_of_complex_words = len(complex_words) / len(words)

fog_index = 0.4 * (average_sentence_length + percentage_of_complex_words)

print(f'Average Sentence Length: {average_sentence_length}')
print(f'Percentage of Complex Words: {percentage_of_complex_words}')
print(f'Fog Index: {fog_index}')
```

```python
import nltk
from nltk.corpus import stopwords
import string

# Loading text from the output.txt file
with open(TextFile, 'r', encoding='utf-8') as file:
    text = file.read()

# Tokenize the text into words
words = nltk.word_tokenize(text)

# Load the NLTK stopwords and create a set of punctuation characters
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)

# Remove stopwords and punctuation from words
cleaned_words = [word for word in words if word.lower() not in stop_words and word not in punctuation]

# Calculate the Word Count after cleaning
word_count_after_cleaning = len(cleaned_words)

# Print the result
print(f'Word Count after cleaning: {word_count_after_cleaning}')
```

## 6.	Syllable Count Per Word

```python
import re

def count_syllables(word):
    # define regular expression pattern
    pattern = re.compile(r'[aeiouyAEIOUY]+')

    # exceptions with es or ed
    if word.endswith(("es", "ed")):
        return max(1, len(pattern.findall(word)) - 1)
    else:
        return max(1, len(pattern.findall(word)))

with open(TextFile, 'r', encoding='utf-8') as file:
    text = file.read()

words = text.split()  # split text into word

syllables_per_word = [count_syllables(word) for word in words]

average_syllables_per_word = sum(syllables_per_word) / len(syllables_per_word)

print(f'Average SYLLABLE PER WORD: {average_syllables_per_word}')

for word, syllable_count in zip(words, syllables_per_word):
    print(f'Word: {word}, Syllables: {syllable_count}')
```

## 7.	Personal Pronouns

```python
import re

# define a regular expression pattern to match personal pronouns  excluding US
pattern = re.compile(r'\b(?!(US)\b)(I|we|my|ours|us)\b', flags=re.IGNORECASE)

with open(TextFile, 'r', encoding='utf-8') as file:
    text = file.read()

personal_pronouns = pattern.findall(text)
personal_pronoun_count = len(personal_pronouns)

print(f'Personal Pronoun Count: {personal_pronoun_count}')
```

## 8.	Average Word Length

```python

with open(TextFile, 'r', encoding='utf-8') as file:
    text = file.read()

words = text.split()

total_characters = sum(len(word) for word in words)
total_words = len(words)

average_word_length = total_characters / total_words

print(f'Average Word Length: {average_word_length}')
```
